```python
# Importing required libraries 
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
```

```python
# Loading DataFrame
df=pd.read_csv('data.csv')
```

```python
# Predictors 
X=df.drop('mh_disorder_current', axis=1)

# Target
y=df['mh_disorder_current']
```

```python
# Explanatory Modelling
model = LogisticRegression()
model.fit(X, y)
```

```python
# To find the coefficent of the model
coef_df = pd.DataFrame({'predictor': X.columns, 'coefficient': model.coef_[0]})
coef_df
```

```python
# To find the top 10 coefficents
coef_df.sort_values(by=['coefficient'], ascending=False).head(10)
```
